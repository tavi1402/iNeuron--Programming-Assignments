# 1. Write a Python program to convert kilometers to miles?

```python

conversion_factor = 0.621371

km = float(input("Enter the kilometer value : "))

miles = km * conversion_factor

print("{} km = {} miles".format(km,miles))
```

# 2. Write a Python program to convert Celsius to Fahrenheit?

```python
def cel_to_Farh():
    celsius = float(input("Enter the temperature in celsius : "))
    farenheit = ((9/5)*celsius)+32
    print('%0.2f째 Celsius is equal to %.2f째 Fahrenheit' %(celsius,farenheit))
    # use shif+option+8 for degree symbol
    
cel_to_Farh()
```

# 3. Write a Python program to display calendar?

```python
import calendar

def display_calender():
    year = int(input("Enter the calender year: "))
    print(calendar.calendar(year))
    
display_calender()
```

# 4. Write a Python program to solve quadratic equation?

```python
import cmath
import math

def quardic_equation(a,b,c):
    try:   
        discriminant = b*b-4*a*c
        sqrtval = math.sqrt(abs(discriminant)) 

        if discriminant > 0:
            root_1 = (-b-sqrtval)/(2 * a)
            root_2 = (-b+sqrtval)/(2 * a)
            print("Real and different roots",root_1,root_2)


        elif discriminant == 0:
            root_1 = -b/(2*a)
            root_2 = -b/(2*a)
            print("Real and same ",root_1,root_2)

        else:
            root_1 = (-b+cmath.sqrt(discriminant))/(2*a)
            root_2 = (-b-cmath.sqrt(discriminant))/(2*a)
            print("complex(imaginary roots)",root_1,root_2)
#     except ZeroDivisionError:
#         print("Zero Division Error occurred")
    except Exception as e:
        print(e)
        
a = int(input('Enter a value: '))
b = int(input('Enter b value: '))
c = int(input('Enter c value: '))

quardic_equation(a,b,c)

    
```

### 1. Write a Python program to check if the given number is a Disarium Number?

```python
def DisariumNumber():
    num = input('Enter a Number: ')
    sum = 0
    for i in range(len(num)):
        sum = sum + int(num[i])**(i+1)
    if sum == int(num):
        print(f'{num} is a Disarium Number')
    else:
        print(f'{num} is a Not Disarium Number') 

DisariumNumber()
DisariumNumber()
DisariumNumber()

```

### 2.Write a Python Program to print all Disarium numbers between 1 to 100 ?

```python
def printDisariumNumbers(start,end):
    output_num = []
    for number in range(start,end+1):
        sum = 0
        for i in range(len(str(number))):
            sum = sum + int(str(number)[i])**(i+1)
        if sum == number:
            output_num.append(number)
    return output_num
            
        
printDisariumNumbers(1,100)
```

### 3.Write a Python Program to check if the given number is Happy Number ?

```python
def is_Happy_num(n):
    past = set()
    while n != 1:
        n = sum(int(i)**2 for i in str(n))
        if n in past:
            return(f'the {n} is not happy number')
        past.add(n)
    return((f'the {n} is a happy number'))
print(is_Happy_num(7))
print(is_Happy_num(33))
print(is_Happy_num(9))
```

### 4.Write a Python Program to print all Happy numbers between 1 and 100 ?

```python
def happy_numbers(n):
    past = set()
    while n != 1:
        n = sum(int(i)**2 for i in str(n))
        if n in past:
            return False
        past.add(n)
    return True
print([x for x in range(101) if happy_numbers(x)][:100])
```

### 5.Write a Python Program to determine whether the given number is a Harshad Number ?


### Harshad number:
- If a number is divisible by the sum of its digits, then it will be known as a Harshad Number.
- For example:
- number 156 is divisible by the sum (12) of its digits (1, 5, 6 ).Some Harshad numbers are 8, 54, 120, etc.



```python
def checkHarshadNum():
    num = input('Enter a Number: ')
    sum = 0
    for i in range(len(num)):
        sum = sum + int(num[i])
    if int(num)%sum == 0:
        print(f'{num} is a Harshad Number')
    else:
        print(f'{num} is a Not Harshad Number')
        
checkHarshadNumber()
checkHarshadNumber()
checkHarshadNumber()

```

### 6.Write a Python Program to print all pronic numbers between 1 and 100 ?

 ##pronic number
